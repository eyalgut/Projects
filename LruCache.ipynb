```python
class Node:
    def __init__(self,k,v):
        self.k=k
        self.v=v
        self.next = None
        self.prev=None   
class List:
    def __init__(self):
        self.head=None
        self.tail=None
    # adds and returns the node
    def append(self,k,v:int):
        if self.tail is None:
            self.head = Node(k,v)
            self.tail = self.head
        else:
            node = Node(k,v)
            self.tail.next= node
            node.prev = self.tail
            self.tail = node
        return self.tail   
    def remove(self,node: Node):
        if node == self.head:
            self.pop()
            return
        prev = node.prev
        nextNode = node.next
        
        if prev is not None:
            prev.next = nextNode
        if nextNode is not None:
            nextNode.prev = prev
        
    def moveNodeToEnd(self,node:Node):
        if node == self.tail:
            return
        
        self.remove(node)
        node.next=None
        self.tail.next = node
        node.prev = self.tail
        self.tail=node
        
    def pop(self):
        if self.head is None:
            return
        t = self.head
        self.head=self.head.next
        return t
    
    def __repr__(self):
        st=""
        t=self.head
        while t is not None:
            st+=str(t.k)+'_'+str(t.v) + " "
            t=t.next
        return st
        
class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.keys=List() # list of nodes
        self.nodes={} # key to node

    def get(self, key: int) -> int:     
        # O(n)
        if key in self.nodes:
            self.keys.moveNodeToEnd(self.nodes[key])
            return self.nodes[key].v
        
        return -1

    def put(self, key: int, value: int) -> None:
        if len(self.nodes) == self.capacity and key not in self.nodes:
            # remove
            del self.nodes[self.keys.pop().k]
            # new
            node = self.keys.append(key,value)
            self.nodes[key]=node
            return
        
        if key in self.nodes:
            self.keys.moveNodeToEnd(self.nodes[key])  
            self.nodes[key].v=value
        else:
            node = self.keys.append(key,value)
            self.nodes[key]=node

    def __repr__(self):
        return str(self.keys) 


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
